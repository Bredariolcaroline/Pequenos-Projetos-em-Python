```python
!pip install yfinance
import yfinance as yf
import matplotlib.pyplot as plt
```

```python
acoes = ['PETR4.SA', 'ABEV','VALE', 'GOOG']
```

```python
dados_acoes = yf.download(acoes,period='1d',start= '2022-01-01')
```

```python
dados_acoes['Close'].plot()
plt.xlabel('Data')
plt.ylabel('Preço de fechamento')
plt.title('Rastreador de ações')
plt.legend(acoes,fontsize=8)

plt.show()
```

```python
#normalizando os preços
dados_normalizados = dados_acoes['Close'] / dados_acoes['Close'].iloc[0]
dados_normalizados.plot()
plt.xlabel('Data')
plt.ylabel('Preço Normalizado')
plt.title('Rastreador de ações')
plt.legend(acoes,fontsize=8)

plt.show()
```
